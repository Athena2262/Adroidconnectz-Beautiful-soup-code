```python
import requests
from bs4 import BeautifulSoup

# Add the url
url = 'https://en.wikipedia.org/wiki/Cat'
```

```python
# Make a request to the website
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```
